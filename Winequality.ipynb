```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
data = pd.read_csv('winequality-red.csv')
```

```python
#Displaying the first five rows of dataset
data.head(5)
```

```python
# Function to create a histogram
def plot_histogram(data, column):
    plt.hist(data[column], bins=10, color='skyblue', edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
# Function to create a scatter plot
def plot_scatter(data, x_column, y_column):
    plt.scatter(data[x_column], data[y_column], color='green', alpha=0.5)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(f'Scatter Plot of {x_column} vs. {y_column}')
    plt.show()
```

```python
# Function to create a heatmap
def plot_heatmap(data):
    plt.figure(figsize=(10,8))
    sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
# Histogram plot between quality and frequency
plot_histogram(data, 'quality')
```

```python
# Scatterplot between alcohol and quality
plot_scatter(data, 'alcohol', 'quality')
```

```python
#Correlation heatmap of the dataset
plot_heatmap(data)
```
